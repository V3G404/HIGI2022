````python
def euler(x0,v0,omega_0,t0=0,dt=0.01,nsteps=100):
  x=[x0]
  t=[t0]
  for i in range(1,nsteps):
    vi=v0-omega_0**2*x0*dt
    xi=x0+v0*dt
    x.append(xi)
    t.append(t0+dt*i)
    x0=xi
    v0=vi
  return x,t
````

````python
x,t=euler(1,0,2,nsteps=1000)
plt.plot(t,x)
````

````python
x,t=euler(0,1,2,nsteps=1000)
plt.plot(t,x)
````

````python
def euler2(x0,v0,omega_0,t10=0,dt1=0.01,nsteps=100):
  x1=[x0]
  t1=[t10]
  for i in range(1,nsteps):
    vi=v0-omega_0**2*np.sin(x0)*dt1
    xi=x0+v0*dt1
    x1.append(xi)
    t1.append(t10+dt1*i)
    x0=xi
    v0=vi
  return x1,t1
````

````python
x1,t1=euler2(1,0,2,nsteps=1000)
plt.plot(t1,x1)
````

````python
x,t=euler(1,0,2,nsteps=1000)
x1,t1=euler2(1,0,2,nsteps=1000)
plt.plot(t,x)
plt.plot(t1,x1)
````
