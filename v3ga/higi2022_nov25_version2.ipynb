# **TRAYECTORIA DE LOS PLANTEAS ⏰**




````python
def sun(x0, y0, z0, vx0, vy0, vz0, t0=0, dt=0.01 ,nsteps=1000):
    G = 4*(np.pi**2)
    M = 1
    ruby = G*M
    rp = np.sqrt( x0**2 + y0**2 + z0**2 ) 
    x = [x0] 
    y = [y0]
    z = [z0]
    t = [t0]
    
    for i in range(1, nsteps):
        Vxi = vx0 - ( (ruby) / rp**3 )*x0 *dt 
        xi = x0 + Vxi*dt
        Vyi = vy0 - ( (ruby) / rp**3 )*y0 *dt 
        yi = y0 + Vyi*dt
        Vzi = vz0 - ( (ruby) / rp**3 )*z0 *dt 
        zi = z0 + Vzi*dt
        x.append(xi)
        y.append(yi)
        z.append(zi)
        t.append(t0+dt*i)
        x0 = xi
        vx0 = Vxi 
        y0= yi
        vy0 = Vyi
        z0 = zi
        vz0 = Vzi
        
    return x, y, z, t
````
